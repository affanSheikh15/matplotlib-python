```python
import matplotlib.pyplot as plt                      
x = [1,2,3,4]
y = [10,20,15,25]

plt.plot(x,y,color = 'green',marker ='o')
plt.title('Dataset of X and Y Axis')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# png = extension , dpi = for resolution clear, bbox_inches = to remove exrta spacing,for_specific folder = folder_name/
plt.savefig('line_plot.png',dpi = 300,bbox_inches = 'tight')
plt.show()
```
