```python
import matplotlib.pyplot as plt         
hours_studies = [1,2,3,4,5,6,7,8,9]
exam_scores = [55,60,65,70,75,80,85,90,95]

plt.scatter(hours_studies,exam_scores, color='green',marker='o',label = 'Student Scores Data')
plt.xlabel("No. of Hours Student Studied")
plt.ylabel("No. of Scores Student Got")
plt.legend()
plt.grid(True)
plt.title("Scatter Graph of Students Data")
plt.show()
```

```python
# COMPARISON BTW TWO DATASETS
import matplotlib.pyplot as plt

plt.scatter([1,2,3],[50,55,60], color='green',marker='^',label = 'Group A')
plt.scatter([1,2,3],[45,52,57], color='blue',marker='^',label = 'Group B')
plt.xlabel("No. of Hours Student Studied")
plt.ylabel("No. of Scores Student Got")
plt.legend()
plt.grid(True)
plt.title("Comparison of Two Classes")
plt.show()
```
