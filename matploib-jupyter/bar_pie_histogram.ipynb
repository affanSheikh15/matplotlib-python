```python
import matplotlib.pyplot as plt              

products = ['A','B','C','D']
sales = [1200,1000,500,800]
# horizontal bhi hostake e.g: plt.barh
plt.bar(products,sales,color = 'orange',label = '2025 Sales')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.legend()
plt.title('Sales Comparison Display')
plt.show()
```

```python
# PIE CHART
import matplotlib.pyplot as plt          

regions = ['North','South','East','West']
revenue = [3000,2000,1500,1000]

plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=['green','red','purple','blue'])
plt.title("Regions and Revenue PieChart")
plt.show()
```

```python
# histogram # 15 ki range per bins

import matplotlib.pyplot as plt             
scores = [46,78,49,58,53,68,64,71,79,82,80,84,95,42,55]
plt.hist(scores,bins=5,color='purple',edgecolor = 'black' )
plt.title("HISTOGRAM OF STUDENT SCORES")
plt.show()
```
