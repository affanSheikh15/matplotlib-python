```python
# PLOT FUNCTIONS (months and sales)
import matplotlib.pyplot as plt

months = [1,2,3,4,5,6]
sales = [10000,40000,35000,50000,60000,55000]

plt.plot(months,sales,color = 'red', linestyle = '--', linewidth = 1 , marker = 'o',label = '2025 Display Sales Chart')
plt.xlabel("Months")
plt.ylabel("No. of Sales")
plt.title("Monthly Sales Data Report")
# CHART KE ANDAR JO TEXT SHOW HORHA WOH UPPER LEFT , LOWER RIGHT 
plt.legend(loc='upper left', fontsize = 10)
# CHART KE ANDAR JO DOTTED LINES OR BOXES AARAHE UNKE LIYE
plt.grid(color = 'gray',linestyle =":",linewidth = 1)
#LIMITS ASSIGN KRTE OR CLEAR UNDERSTANDING OR VIEW KE LIYE GRAPH KI
plt.xlim(1,6)
plt.ylim(0,62000)
plt.xticks([1,2,3,4,5,6],['M1','M2','M3','M4','M5','M6'])
plt.show()
```
