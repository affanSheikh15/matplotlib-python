```python
import matplotlib.pyplot as plt               
x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1)#1 row,2 column,1 subplot
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2)#1 row , 2 column, 2 subplot
plt.bar(x,y)
plt.title('Barchart')

plt.show()
```

```python
# PROFESSIONAL WAY 
import matplotlib.pyplot as plt               

fig, ax = plt.subplots(1,2, figsize = (10,5))

x = [1,2,3,4]
y = [10,20,15,25]


ax[0].plot(x,y , color = 'blue')
ax[0].set_title('Line Chart')

ax[1].bar(x,y,color = 'green')
ax[1].set_title('BarChart')

plt.suptitle('COMPARISON BTW LINE CHART AND BAR CHART')
plt.tight_layout()
plt.show()
```
